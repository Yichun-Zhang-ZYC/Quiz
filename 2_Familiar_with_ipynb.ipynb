## Visualize Wine Class Distribution
We create a bar chart showing how many samples belong to each wine class (1, 2, and 3).

Each class corresponds to a different grape cultivar.  
This gives us an overview of how balanced the dataset is across wine types.


```python
import pandas as pd
```

```python
wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
wine.head()
```

```python
import matplotlib.pyplot as plt

# Add column names for clarity (class is the first column)
columns = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
           'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
           'Color intensity', 'Hue', 'OD280/OD315', 'Proline']
wine.columns = columns

# Count how many samples per wine class
class_counts = wine['Class'].value_counts().sort_index()

# Plot using matplotlib
plt.figure(figsize=(6,4))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Samples per Wine Class')
plt.xlabel('Wine Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
